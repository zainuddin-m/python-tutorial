#Build a guessing game

```python
#guessing game which asks the player to guess the secret word
secret_word = "giraffe"
guess = ""

while guess != secret_word:
  guess = input("Enter guess: ")

print("You win!")
```

```python
#adaptation of the guessing game above which allows for only 3 tries using a guess counter
secret_word = "giraffe"
guess = ""
guess_count = 0
guess_limit = 3
out_of_guesses = False

while guess != secret_word and (out_of_guesses == False):
  if guess_count < guess_limit:
    guess = input("Enter guess: ")
    guess_count += 1
  else:
    out_of_guesses = True

if out_of_guesses:
  print("Out of guesses, YOU LOSE!")
else:
  print("You win!")
```

#TRY EXCEPT
used for catching errors in python

```python
#a try except block allows you to try a block of code and if it works then we are great!
#but if it doesnt work then we can account for it

number = int(input("Enter a number: "))
print(number)
#if a user enters not an integer here, the entire program will break

try:
  number = int(input("Enter a number: "))
  print(number)
except:
  print("Invalid Input")
```
